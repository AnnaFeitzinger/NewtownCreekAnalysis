```r
data=read.csv("/Users/dnalc/Documents/CUNY Newton Creek Diversity/Data/abundance_table_genus.tsv", sep="\t", header=T,row.names = 'tax')
data = data[,c(1:8)]
```

```r
#Subset table to your species of interest
```

```r
counts = data[grep("Enterobacteriaceae;Enterobacter", rownames(data)),]
```

```r
# Plot number of DNA sequences found for that species at the different site locations
```

```r
df <- data.frame(reads=as.numeric(as.vector(counts[1,])),
                 site=c("Site 1", "Site 2", "Site 3", "Site 4", "Site 5", "Site 6", "Site 7","Site 8"))

p<-ggplot(data=df, aes(x=site, y=reads)) +
  geom_bar(stat="identity", fill="steelblue")+
  theme_minimal() +
  labs(title="Enterobacter", 
       x="Collection Site", y = "Counts")
p
```
