### Plotting Tutorial in R
R is a powerful language for statistical computing and graphics. It provides many functions for creating various types of plots, from simple scatterplots to complex visualizations. In this tutorial, we'll cover the basics of plotting in R, including how to use both the base R plotting system and the ggplot2 package.

#### Base R Plotting
The base R plotting system provides a set of functions for creating plots. The primary function is plot(), which can create different types of plots depending on

#### Creating a Simple Scatterplot
To create a scatterplot of two numeric variables, use the plot() function with the variables as arguments:

```r
# Create some sample data
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 1, 3, 5)

# Create a scatterplot
plot(x, y)
```

### Adding Labels and Titles
You can add labels to the x-axis and y-axis using the xlab and ylab arguments, and a title to the plot using the main argument:

```r
plot(x, y, xlab = "X-axis Label", ylab = "Y-axis Label", main = "My Scatterplot")
```

### Changing the Plot Type
#### The plot() function can create different types of plots depending on the type argument. Some common values for type are:

"p" for points (the default)

"l" for lines

"b" for both points and lines

"h" for histogram-like vertical lines

#### Customizing Appearance
You can customize the appearance of your plots using various graphical parameters. Some common parameters include:

col: Color of points or lines

pch: Symbol to use for points

cex: Size of points

lty: Line type

lwd: Line width

```r
# Create a line plot
plot(x, y, type = "l", xlab = "X-axis", ylab = "Y-axis", main = "Line Plot")

```

### Other Common Base R Plots
Bar charts: barplot()

Histograms: hist()

Boxplots: boxplot()

### ggplot2
ggplot2 is a powerful and flexible package for creating plots in R. It is based on the grammar of graphics, which provides a structured way to describe the components of a plot.

#### Installing ggplot2
If you haven't already, install ggplot2 by running:

```r
install.packages("ggplot2")

library(ggplot2)
```

### Creating a ggplot
A ggplot2 plot is created using the ggplot() function. You specify the data frame to use and the variables to map to different aesthetic elements of the plot (e.g., x-axis, y-axis, color).

```r
# Create a ggplot object
p <- ggplot(data.frame(x, y), aes(x = x, y = y))
```

### Adding Geoms
Geoms are the geometric objects that represent the data in a plot. For example, you can use geom_point() to create a scatterplot, geom_line() to create a line plot, or geom_bar() to create a bar chart.

```r
# Add points to the plot
p + geom_point()
```

### Adding Labels and Titles
You can add labels and titles to a ggplot2 plot using the labs() function:

```r
# Add labels and a title
p +
    geom_point() +
    labs(x = "X-axis Label", y = "Y-axis Label", title = "My ggplot2 Plot")
```

### Customizing Appearance
ggplot2 provides many options for customizing the appearance of your plots. You can change colors, sizes, fonts, and more using various functions and arguments.



```r
# Customize the appearance of the plot
p +
    geom_point(color = "red", size = 3) + # Color and size
    labs(x = "X-axis", y = "Y-axis", title = "Custom ggplot2 Plot") +
    theme_minimal() # Use the minimal theme
```

### Different Types of ggplot2 Plots
Scatterplots: geom_point()

Line plots: geom_line()

Bar charts: geom_bar() or geom_col()

Histograms: geom_histogram()

Boxplots: geom_boxplot()

#### Faceting
Faceting is the process of creating multiple small plots based on the values of one or more categorical variables. You can use the facet_wrap() or facet_grid() functions to create faceted plots in ggplot2.

#### Themes
ggplot2 comes with several built-in themes that control the overall appearance of the plot. You can also create your own custom themes.



```r
# Use the 'theme_bw' theme
ggplot(data.frame(x, y), aes(x = x, y = y)) +
    geom_point() +
    theme_bw()
```

### Example: Creating a Bar Chart with ggplot2
Let's create a bar chart using the ggplot2 package. First, we'll create some sample data:

```r
# Create some sample data
categories <- c("A", "B", "C", "D")
values <- c(10, 25, 15, 30)
data <- data.frame(categories, values)

### Now, we'll create a bar chart using ggplot2:

# Create a bar chart
ggplot(data, aes(x = categories, y = values)) +
    geom_bar(stat = "identity", fill = "steelblue") + # Create the bars, set color
    labs(x = "Category", y = "Value", title = "Bar Chart") + # Add labels
    theme_minimal() # Use a minimal theme
```

In this example, geom_bar(stat = "identity") tells ggplot2 to use the values in the values column as the heights of the bars. The fill argument sets the color of the bars.

## Lets make graphs of our data!

```r
# Create a sample data frame
site_data <- data.frame(
  Site = c("A", "B", "C", "D", "E"),
  Number_of_Genes = c(120, 150, 80, 200, 95),
  Wet_Dry = c('Wet','Dry','Wet','Dry','Wet')
)

# Print the data frame
print(site_data)
```

```r
barplot(site_data$Number_of_Genes, names.arg = site_data$Site,
        xlab = "Site", ylab = "Number of Genes",
        main = "Number of Genes per Site", col = "skyblue")
```

```r
library(ggplot2)
ggplot(site_data, aes(x = Site, y = Number_of_Genes, fill = Wet_Dry)) +
  geom_bar(stat = "identity") +
  xlab("Site") +
  ylab("Number of Genes") +
  ggtitle("Number of Genes per Site, Colored by Wet/Dry") +
  theme_minimal()
```
